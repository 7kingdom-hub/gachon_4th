```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras

file_path = 'C:/Users/7kingdom/Documents/인공지능개론/src/iris/iris.data'

#class = Iris Setosa, Iris Versicolour, Iris Virginica
column_names = ["sepal length in cm", "sepal width in cm", "petal length in cm", "petal width in cm", "class"]          

df = pd.read_csv(file_path, header=None, names=column_names)

df.head(10)
```

```python
X=df.drop('class',axis=1)
X.head()
```

```python
y=df['class']
y.value_counts()
```

```python
Y = pd.get_dummies(y).values
```

```python
X = X.values
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=0) 
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

<h3> 뉴런 수 정하기 </h3>
    

자 X의 피쳐가 4개니까 shape를 4로 설정하고
y의 클래스가 3개니까 최종 출력층을 3으로 설정해야한다.

클래스가 여러개니 다중 분류 손실함수를 사용한다.

또한, 역시 다중 클래스 분류 문제이니 sigmoid 대신 softmax를 사용한다.

```python
<h3>퍼셉트론 해보기</h3>

책에 나와있는 사이킷런에서 제공하는 퍼셉트론을 써봤는데

iris 데이터의 출력 클래스가 3개라 선형 모델인 단일 퍼셉트론은 안됀다.

그래서 다중 퍼셉트론으로 레이어를 3개 놔서 한 번 해보자
```

```python
#Defining the model 

from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD,Adam

#퍼셉트론 가져오기
from sklearn.linear_model import Perceptron

#다중 퍼셉트론 가져오기
from sklearn.neural_network import MLPClassifier

# MLP 모델 생성
mlp_model = MLPClassifier(hidden_layer_sizes=(10, 8, 6), activation='tanh', max_iter=1000, random_state=42)

# 모델 학습
mlp_model.fit(X_train, y_train)

# 모델 평가
accuracy = mlp_model.score(X_test, y_test)
print("Accuracy:", accuracy)
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras

file_path = 'C:/Users/7kingdom/Documents/인공지능개론/src/iris/iris.data'

#class = Iris Setosa, Iris Versicolour, Iris Virginica
column_names = ["sepal length in cm", "sepal width in cm", "petal length in cm", "petal width in cm", "class"]          

df = pd.read_csv(file_path, header=None, names=column_names)

df.head(10)
```

```python
X=df.drop('class',axis=1)
X.head()
```

```python
y=df['class']
y.value_counts()
```

```python
Y = pd.get_dummies(y).values
```

```python
X = X.values
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=0) 
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

<h3> 뉴런 수 정하기 </h3>
    

자 X의 피쳐가 4개니까 shape를 4로 설정하고
y의 클래스가 3개니까 최종 출력층을 3으로 설정해야한다.

클래스가 여러개니 다중 분류 손실함수를 사용한다.

또한, 역시 다중 클래스 분류 문제이니 sigmoid 대신 softmax를 사용한다.

```python
#Defining the model 

from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD,Adam


model = Sequential()

model.add(Dense(10,input_shape=(4,),activation='tanh'))
model.add(Dense(8,activation='tanh'))
model.add(Dense(6,activation='tanh'))
model.add(Dense(3,activation='softmax'))

model.compile(Adam(lr=0.04),'categorical_crossentropy',metrics=['accuracy'])

model.summary()
```

```python
model_history=model.fit(x=X_train, y=y_train, epochs=30, batch_size=32,validation_data= (X_test,y_test))
y_pred = model.predict(X_test)

y_test_class = np.argmax(y_test,axis=1)
y_pred_class = np.argmax(y_pred,axis=1)
```

```python
#Accuracy of the predicted values
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test_class,y_pred_class))
print(confusion_matrix(y_test_class,y_pred_class))
```

```python
import numpy as np

import pandas as pd

import seaborn as sns
```

```python
file_path = 'C:/Users/7kingdom/Documents/인공지능개론/src/iris/iris.data'

#class = Iris Setosa, Iris Versicolour, Iris Virginica
column_names = ["sepal length in cm", "sepal width in cm", "petal length in cm", "petal width in cm", "class"]          

df = pd.read_csv(file_path, header=None, names=column_names)

df.head(10)
```

```python
X=df.drop('class',axis=1)
X.head()
```

```python
y=df['class']
y.head(10)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
iris_ML = LogisticRegression(random_state=0)
iris_ML.fit(X_train, y_train)

pred_iris_ML = iris_ML.predict(X_test)

print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(y_test, pred_iris_ML))
print (confusion_matrix(y_test, pred_iris_ML))
```

위에 처럼 오류가 발생한다. 한 번 해결해보자,


  ChatGPT에게 물어보니 Logistic Regression 모델의 lbfgs 최적화 알고리즘을 사용하는데 이게 최대 반복 횟수까지 도달해서 발생하는 문제였다.
  
  해결방법
  
  1. max_iter을 증가시킨다.<br>
      iris_ML = LogisticRegression(random_state=0, max_iter=1000)  # 예시로 1000으로 증가시킴<br><br>
  2. 다른 최적화 알고리즘을 사용한다.<br>
      알고리즘 관련 매개변수는 solver인데 이를 newton-cg을 사용하든 sag를 사용하든 다른 걸 한 번 써본다.<br>
      iris_ML = LogisticRegression(random_state=0, solver='sag')

  

```python
iris_ML = LogisticRegression(random_state=0, max_iter=1000)
iris_ML.fit(X_train, y_train)

pred_iris_ML = iris_ML.predict(X_test)

print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(y_test, pred_iris_ML))
print (confusion_matrix(y_test, pred_iris_ML))
```
